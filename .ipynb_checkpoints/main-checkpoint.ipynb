### Some feature engineering to do more analysis

```python
import pandas as pd

# Calculate Employment Growth Rate
data['employment_growth_rate'] = data['all'].pct_change() * 100

# Calculate Age Group Averages
age_groups = ['16-24', '25-54', '55-64', '65+']
for age_group in age_groups:
    data[f'avg_employment_{age_group}'] = data[[f'{gender}_{age_group}' for gender in ['women', 'men']]].mean(axis=1)

# Calculate Gender Gap
data['gender_gap'] = data['women'] - data['men']

# Calculate Educational Attainment Gap
education_levels = ['less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']
for education_level in education_levels:
    data[f'education_gap_{education_level}'] = data[f'women_{education_level}'] - data[f'men_{education_level}']

# Display the modified dataset
data.head()
```

## ML models:


### 1/ Cluster Analysis. 


```python
# Select the columns of interest from your DataFrame
selected_columns = data[['year', '16-24', '25-54', '55-64', '65+', 'women', 'men', 'black', 'hispanic', 'white']]

# Standardize the data
scaler = StandardScaler()
data_standardized = scaler.fit_transform(selected_columns)

inertia = []

# Try different values of K from 1 to a reasonable maximum
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_standardized)
    inertia.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)

cluster_labels = kmeans.fit_predict(data_standardized)  # Fit and predict in one step

```

```python
#with dimension reduction: 
# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data_standardized)

#plot
plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=cluster_labels, cmap='viridis', alpha=0.7)
plt.title('K-means Clustering Result')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

### 2/ ARIMA model:

```python
years = data['year']
employment = data['all']

# Create a line plot
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

plt.plot(years, employment, marker='o', color='blue', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Employment')
plt.title('General Employment per Year')

plt.grid(True)  # Add grid lines
plt.show()
```

```python
data['year'] = pd.to_datetime(data['year'], format='%Y')
data.set_index('year', inplace=True)
```

```python
train_data = data[:'2017']
test_data = data['2018':'2022']
```

```python
train_arima = train_data['all']
test_arima = test_data['all']
```

```python
column_of_interest = 'all'
time_series_data = data[column_of_interest]
def difference_stationarity_test(data, max_diff=5, alpha=0.05):
    diff_order = 0
    p_value = 1.0
    
    while p_value > alpha and diff_order < max_diff:
        diff_data = data.diff().dropna()
        _, p_value, _, _, _, _ = adfuller(diff_data)
        
        if p_value > alpha:
            diff_order += 1
            data = diff_data
    
    return data, diff_order
```

```python

# Apply differencing until seasonality is removed
stationary_data, diff_order = difference_stationarity_test(time_series_data, max_diff=5)
print(f'Differencing Order for {column_of_interest}: {diff_order}')
```

```python
# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(stationary_data, lags=40, ax=plt.gca())
plt.subplot(212)
plot_pacf(stationary_data, lags=10, ax=plt.gca())
plt.show()
```

```python
history = [x for x in train_arima]
y = test_arima
# make first prediction
predictions = list()
model = ARIMA(history, order=(1,0,2))
model_fit = model.fit()
yhat = model_fit.forecast()[0]
predictions.append(yhat)
history.append(y[0])
```

```python
# rolling forecasts
for i in range(1, len(y)):
    # predict
    model = ARIMA(history, order=(1,0,2))
    model_fit = model.fit()
    yhat = model_fit.forecast()[0]
    # invert transformed prediction
    if i < len(y):
        predictions.append(yhat)
    # observation
    obs = y[i]
    history.append(obs)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

# report performance
mse = mean_squared_error(y, predictions)
print('MSE: '+str(mse))
mae = mean_absolute_error(y, predictions)
print('MAE: '+str(mae))
rmse = math.sqrt(mean_squared_error(y, predictions))
print('RMSE: '+str(rmse))
```
